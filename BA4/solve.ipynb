```python
def get_kmers(seq, k):
    kmers_list = []
    for i in range(len(seq)-k+1):
        kmers_list.append(seq[i:i+k])
    return kmers_list

def trans_dna_to_rna(seq):
    trans = {
        'A':'U',
        'T':'A',
        'C':'G',
        'G':'C'
    }
    return ''.join([trans[i] for i in list(seq)])

def get_rna(seq):
    return ''.join([i if i!='T' else 'U' for i in list(seq)])

def find_all_proper_kmers(seq, acids_string, k):
    kmers_list = get_kmers(seq, k)
    results_list = []
    for kmer in kmers_list:
        if trans_rna_to_amino_acids(get_rna(kmer)) == acids_string or trans_rna_to_amino_acids(trans_dna_to_rna(kmer[::-1])) == acids_string:
            results_list.append(kmer)
            
    return results_list

with open('rosalind_ba4b.txt') as fp:
    lines = fp.readlines()
    seq = lines[0].strip()
    acids_string = lines[1].strip()
    results_list = find_all_proper_kmers(seq, acids_string, 3*len(acids_string))
    for i in results_list:
        print(i)
```

```python
acid_mass = {'G': 57, 'A': 71, 'S': 87, 'P': 97, 'V': 99,
                  'T': 101, 'C': 103, 'I': 113, 'L': 113, 'N': 114,
                  'D': 115, 'K': 128, 'Q': 128, 'E': 129, 'M': 131,
                  'H': 137, 'F': 147, 'R': 156, 'Y': 163, 'W': 186}

def get_all_cycle_kmers(seq, k):
    seq_new = seq + seq[:k-1]
    kmers_list = [seq_new[i:i+k] for i in range(len(seq))]
    return list(kmers_list) 

def calculate_all_possible_mass(seq):
    all_possible_combinations = []
    for i in range(2, len(seq)):
        all_possible_combinations += get_all_cycle_kmers(seq, i)
    all_possible_combinations.append(seq)
    if len(seq) >= 2:
        all_possible_combinations += list(seq)
    all_possible_mass = []
    for acid_tuple in all_possible_combinations:
        all_possible_mass.append(calculate_mass(acid_tuple))
    all_possible_mass.append(0)
    all_possible_mass = sorted(all_possible_mass)
    return all_possible_mass

def calculate_mass(acid_combination):
    mass = 0
    for acid in acid_combination:
        mass += acid_mass[acid]
    return mass

with open('rosalind_ba4c.txt') as fp:
    seq = fp.readline().strip()
    
result = calculate_all_possible_mass(seq)
result = [str(i) for i in result]
print(' '.join(result))
 
```

```python
acid_mass = {'G': 57, 'A': 71, 'S': 87, 'P': 97, 'V': 99,
                  'T': 101, 'C': 103, 'I': 113, 'L': 113, 'N': 114,
                  'D': 115, 'K': 128, 'Q': 128, 'E': 129, 'M': 131,
                  'H': 137, 'F': 147, 'R': 156, 'Y': 163, 'W': 186}

def expand(peptides, acids, max_spectrum):
    expand_result = []
    for peptide in peptides:
        for acid in acids:
            if calculate_mass(peptide + acid) <= max_spectrum:
                expand_result.append(peptide + acid)
    return expand_result

def is_consistent(peptide, spectrum):
    peptide_spectrum = []
    prefix_mass = [0]
    for aa in peptide:
        prefix_mass.append(prefix_mass[-1] + acid_mass[aa])
    for i in range(len(prefix_mass)):
        for j in range(i + 1, len(prefix_mass)):
            peptide_spectrum.append(prefix_mass[j] - prefix_mass[i])
    return all(peptide_spectrum.count(m) <= spectrum.count(m) for m in set(peptide_spectrum))

def iter_peptides_cycle(spectrum, acids):
    peptides = ['']
    target_peptides = []
    max_spectrum_mass = max(spectrum)
    while len(peptides) != 0:
        peptides = expand(peptides, acids, max_spectrum_mass)
        peptides_copy = peptides[:]
        for peptide in peptides:
            mass = calculate_mass(peptide)
            therotial_spectrum = calculate_all_possible_mass(peptide)
            if mass == max_spectrum_mass:
                if therotial_spectrum == spectrum:
                    target_peptides.append(peptide)
                    peptides_copy.remove(peptide)
            elif not is_consistent(peptide, spectrum):
                peptides_copy.remove(peptide)
        peptides = peptides_copy
    return target_peptides

def trans_peptide_to_mass(peptide, acid_mass):
    return '-'.join([str(acid_mass[acid]) for acid in peptide])

with open('rosalind_ba4e.txt') as fp:
    spectrum = list(map(int, fp.readline().strip().split(' ')))
acids = list(acid_mass.keys())

result = iter_peptides_cycle(spectrum, acids)
mass_result = [trans_peptide_to_mass(peptide, acid_mass) for peptide in result]
print(' '.join(sorted(set(mass_result))))
```

```python
def calculate_and_select(peptides, spectrum, N):
    sorted_peptides = sorted(peptides, key=lambda x: calculate_score(x, spectrum), reverse=True)
    if len(sorted_peptides) <= N:
        return sorted_peptides
    else:
        return sorted_peptides[:N]

def iter_leaderboard(spectrum, N, acids):
    leaderboard = ['']
    leaderpeptide = ''
    max_spectrum = max(spectrum)
    while len(leaderboard) != 0:
        leaderboard = expand(leaderboard, acids, max_spectrum)
        for peptide in leaderboard[:]:
            if calculate_mass(peptide) == max_spectrum:
                if calculate_score(peptide, spectrum) > calculate_score(leaderpeptide, spectrum):
                    leaderpeptide = peptide
            elif calculate_mass(peptide) > max_spectrum:
                leaderboard.remove(peptide)
        leaderboard = calculate_and_select(leaderboard, spectrum, N)
    
    return leaderpeptide

with open('rosalind_ba4g.txt') as fp:
    lines = [line.strip() for line in fp.readlines()]
    N = int(lines[0])
    spectrum = list(map(int, lines[1].split(' ')))
    
print(trans_peptide_to_mass(iter_leaderboard(spectrum, N, acids), acid_mass))
```

```python
from collections import Counter

def get_all_pairs(spectrum):
    all_pairs = []
    for i in range(len(spectrum)-1):
        for j in range(i + 1, len(spectrum)):
            if spectrum[j] - spectrum[i] > 0:
                all_pairs.append((spectrum[i], spectrum[j]))
    
    return all_pairs

def spectrum_convolution(spectrum):
    all_pairs = get_all_pairs(spectrum)
    diff_value =list(map(lambda x: x[1]-x[0], all_pairs))
    diff_count = list(Counter(diff_value).items())
    sorted_diff_count = sorted(diff_count, key=lambda x:(x[1], -x[0]), reverse=True)
    result = []
    for diff in sorted_diff_count:
        result += [diff[0]] * diff[1]
    return diff_value, result

with open('rosalind_ba4h.txt') as fp:
    spectrum = sorted(list(map(int, fp.readline().split(' '))))

print(' '.join(list(map(str, spectrum_convolution(spectrum)[1]))))
```

```python
#!/public1/user/wangjiex/software/Python3.9/bin/python3.9
# encoding: utf-8
# Author: Wang Jiexin
# Contact: wangjx68@mail2.sysu.edu.cn
# 2022.06.08

def spectrum_to_convolution_alphat(spectrum, m):
    spectrum = sorted(spectrum)
    convolution = {}
    for i in range(len(spectrum)):
        for j in range(i, len(spectrum)):
            if spectrum[i] < spectrum[j] and \
                    spectrum[j] - spectrum[i] in range(57, 201):
                if spectrum[j]-spectrum[i] in convolution.keys():
                    convolution[spectrum[j]-spectrum[i]] += 1
                else:
                    convolution[spectrum[j]-spectrum[i]] = 1
    mass = sorted(convolution.items(), key=lambda x: (-x[1], x[0]))
    index = m + 1
    ans = []
    for i in range(m+1, len(mass)):
        if mass[i][1] == mass[m][1]:
            index = i + 1
    for i in range(index):
        ans.append(mass[i][0])
    return ans

def peptide_to_spectrum(peptide, is_circle=False):
    spectrum = [0, sum(peptide)]
    if is_circle:
        for i in range(len(peptide)):
            for j in range(i+1, len(peptide)):
                spectrum.append(sum(peptide[i:j]))
            for j in range(0, i):
                spectrum.append(sum(peptide[i:]+peptide[:j]))
    else:
        for i in range(len(peptide)):
            for j in range(i+1, len(peptide)):
                spectrum.append(sum(peptide[i:j]))
    return sorted(spectrum)

def peptide_spectrum_score(peptide, ideal_spectrum, is_circle=False):
    spectrum = peptide_to_spectrum(peptide, is_circle)
    score, i = 0, 0
    local_spectrum = ideal_spectrum[:]
    while i < len(spectrum):
        if spectrum[i] in local_spectrum:
            score += 1
            local_spectrum.remove(spectrum[i])
            i += 1
        else:
            i += 1
    return score

def is_sublist(lsa, lsb):
    # 在使用方法操作的时候要注意是否能够更改原变量
    local_lsa = lsa[:]
    local_lsb = lsb[:]
    while local_lsa:
        value = local_lsa.pop()
        try:
            # 这里这么写会导致输入的spectrum被修改lsb.remove(value)
            local_lsb.remove(value)
        except ValueError:
            return False
    return True

def trim_candidate_to_leaderboard(n, candidate, leaderboard):
    candidate = sorted(candidate,
                       key=lambda x: peptide_spectrum_score(x, ideal_spectrum),
                       reverse=True)
    index = n - 1
    for i in range(n, len(candidate)):
        if peptide_spectrum_score(candidate[i], ideal_spectrum) == \
                peptide_spectrum_score(candidate[n - 1], ideal_spectrum):
            index = i
        else:
            break
    leaderboard = candidate[:index + 1]
    return candidate, leaderboard

def leaderboard_cycle_peptide_sequencing(n, ideal_spectrum, mass_list):
    max_mass = max(ideal_spectrum)
    leaderboard, candidate = [], [[i] for i in mass_list]
    # 初始化leaderboard
    candidate, leaderboard = trim_candidate_to_leaderboard(n, candidate, leaderboard)
    # 判断这一轮的延伸是否能够产生新的序列，作为是否继续延伸-剪枝的策略
    while True:
        # 分支: 不需要删除原有的分支
        candidate = leaderboard[:]
        j, expand = 0, 0
        while j < len(candidate):
            c = candidate[j]
            for m in mass_list:
                tmp = c + [m]
                # 这里如果超过理想谱最大质量则不能加入
                if sum(tmp) <= max_mass:
                    j += 1
                    expand += 1
                    candidate.insert(j, tmp)
            j += 1
        # 更新leaderboard
        deletions = []
        for c in candidate:
            if not is_sublist(c, ideal_spectrum):
                expand -= 1
                deletions.append(c)
        for d in deletions:
            candidate.remove(d)
        candidate, leaderboard = trim_candidate_to_leaderboard(n, candidate, leaderboard)
        leaderboard = sorted(leaderboard,
                            key=lambda x: peptide_spectrum_score(x, ideal_spectrum, is_circle=True),
                            reverse=True)
        if expand == 0:
            return leaderboard[0]

with open('rosalind_ba4i.txt', 'r') as fbj:
    lines = fbj.readlines()
    m, n = int(lines[0].strip()), int(lines[1].strip())
    ideal_spectrum = list(map(int, lines[2].strip().split(' ')))
    mass_list = spectrum_to_convolution_alphat(ideal_spectrum, m)
    print('-'.join(list(map(str, leaderboard_cycle_peptide_sequencing(n, ideal_spectrum, mass_list)))))

    
```

```python
import numpy as np

def calculate_prefix_mass(peptide, acid_mass):
    acids_list = list(peptide)
    mass_list = np.array([acid_mass[acid] for acid in acids_list])
    prefix_mass = np.cumsum(mass_list)
    
    return [0] + prefix_mass.tolist()

def get_linear_spectrum(peptide, acid_mass):
    prefix_mass = calculate_prefix_mass(peptide, acid_mass)
    linear_spectrum = []
    for i in range(len(peptide)):
        for j in range(i+1, len(peptide)+1):
            linear_spectrum.append(prefix_mass[j] - prefix_mass[i])
    return [0] + sorted(linear_spectrum)

with open('rosalind_ba4j.txt') as fp:
    peptide = fp.readline().strip()
prefix_mass = calculate_prefix_mass(peptide, acid_mass)
linear_spectrum = get_linear_spectrum(peptide, acid_mass)
print(' '.join(list(map(str,linear_spectrum))))
```

```python
from collections import Counter

def Turnpike(L):
    max_A = max(L)
    L = [i for i in L if i > 0]
    L.remove(max_A)
    A = [0, max_A]
    L_counter = Counter(L)
    A_set = set([0, max_A])
    result = Reconstruct(L_counter, A_set, max_A)
    if result:
        A = sorted(result)
        return A
    else:
        print("无解")
        return None

def get_possible_distances(A_set, new_point):
    return [abs(new_point - point) for point in A_set]

def Reconstruct(L_counter, A_set, max_A):
    if not L_counter:
        return A_set
    d = max(L_counter)
    candidates = [d, max_A - d]
    for new_point in candidates:
        if new_point in A_set:
            continue
        distances = get_possible_distances(A_set, new_point)
        distances_counter = Counter(distances)
        if not distances_counter <= L_counter:
            continue
        # 更新 L_counter 和 A_set
        L_counter_new = L_counter - distances_counter
        A_set_new = A_set.copy()
        A_set_new.add(new_point)
        # 递归调用
        result = Reconstruct(L_counter_new, A_set_new, max_A)
        if result:
            return result
    return None

with open('rosalind_ba4m.txt') as fp:
    L = list(map(int, fp.readline().split(' ')))
    
print(' '.join(map(str, Turnpike(L))))    
```
