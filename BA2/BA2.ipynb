```python
# Function to find all k-mers in a sequence
def find_all_kmers(seq, k):
    kmer_result = []
    for i in range(len(seq) - k + 1):
        temp = seq[i:i+k]
        if temp not in kmer_result:
            kmer_result.append(temp)
    return kmer_result

# Function to calculate base percentages in a sequence
def calculate_percentage(seq):
    bases = ["A", "C", "G", "T"]
    count_result = []
    for base in bases:
        count_result.append(seq.count(base) / len(seq))
    return count_result

# Optimized profile creation without pseudocounts
def make_profile(seq_list):
    """ Return a profile dictionary without pseudocounts """
    k = len(seq_list[0])
    t = len(seq_list)
    
    # Initialize profile dictionary
    profile = {'A': [0] * k, 'C': [0] * k, 'G': [0] * k, 'T': [0] * k}
    
    # Count the occurrences of each base at each position
    for seq in seq_list:
        for i, base in enumerate(seq):
            profile[base][i] += 1
    
    # Convert counts to frequencies
    for base in profile:
        profile[base] = [count / t for count in profile[base]]
    
    return profile

# Function to calculate the probability of a k-mer given a profile
def calcualte_probility(kmer, profile):
    prob = 1
    for i, base in enumerate(kmer):
        prob *= profile[base][i]  # Use profile[base][i] to get the probability
    return prob

# Function to find the best k-mer in a sequence based on a profile
def find_best_kmer(seq, profile, k):
    kmer_list = find_all_kmers(seq, k)
    
    # Calculate probabilities for all k-mers
    probility_result = [calcualte_probility(kmer, profile) for kmer in kmer_list]
    
    # Find the k-mer with the highest probability
    max_prob = max(probility_result)
    for i, prob in enumerate(probility_result):
        if prob == max_prob:
            return kmer_list[i]

# Function to find the base with the highest frequency in a list
def find_base(list):
    bases = ["A", "C", "G", "T"]
    for index in range(len(list)):
        if list[index] == max(list):
            return bases[index]

# Function to calculate the Hamming distance between two sequences
def Hamming_distance(seq1, seq2):
    compare_result = len([i for i in range(len(seq1)) if seq1[i] != seq2[i]])
    return compare_result

# Function to calculate the score of a set of motifs
def score(motif):
    motif = [[single_base for single_base in seq] for seq in motif]
    motif_trans = list(zip(*motif))
    motif_trans = ["".join(list(seq)) for seq in motif_trans]
    
    count_result = [calculate_percentage(seq) for seq in motif_trans]
    
    base_most = [find_base(position_result) for position_result in count_result]
    
    most_seq = "".join(base_most)
    
    # Calculate the total score based on Hamming distance to the consensus sequence
    score = sum([Hamming_distance(seq_item, most_seq) for seq_item in motif])
    
    return score

# Main function to run the Greedy Motif Search algorithm
def main():
    with open("rosalind_ba2d.txt") as fp:
        result = fp.readlines()
        k = int(result[0].split(" ")[0])
        t = int(result[0].split(" ")[1].strip())
        seqs = [seq.strip() for seq in result[1:]]
    
    # Initialize best motifs as the first k-mer of each sequence
    best_motifs = [seq[0:k] for seq in seqs]
    
    # Find all k-mers in the first sequence
    first_kmers = find_all_kmers(seqs[0], k)
    
    for kmer1 in first_kmers:
        motif = [kmer1]
        profile = make_profile(motif)  # Create the profile for the first motif
        for i in range(1, t):
            motif.append(find_best_kmer(seqs[i], profile, k))  # Find the best k-mer for each sequence
            profile = make_profile(motif)  # Update the profile with the new motifs
        if score(motif) < score(best_motifs):
            best_motifs = motif
    
    # Output the best motifs
    for motif in best_motifs:
        print(motif)

# Run the main function
if __name__ == "__main__":
    main()
```

```python
import random

# 构建带伪计数的Profile矩阵
def make_profile_with_pseudocounts(motifs):
    t = len(motifs)
    k = len(motifs[0])
    profile = {'A': [1] * k, 'C': [1] * k, 'G': [1] * k, 'T': [1] * k}
    
    for motif in motifs:
        for i, base in enumerate(motif):
            profile[base][i] += 1
    
    for base in profile:
        profile[base] = [count / (t + 4) for count in profile[base]]
    
    return profile

# 计算给定k-mer的概率
def calculate_probability(kmer, profile):
    prob = 1
    for i, base in enumerate(kmer):
        prob *= profile[base][i]
    return prob

# 从序列中找到所有可能的k-mers
def find_all_kmers(seq, k):
    return [seq[i:i + k] for i in range(len(seq) - k + 1)]

# 基于Profile随机选择k-mer
def profile_randomly_generated_kmer(dna, k, profile):
    kmer_list = find_all_kmers(dna, k)
    probabilities = [calculate_probability(kmer, profile) for kmer in kmer_list]
    total_prob = sum(probabilities)
    
    probabilities = [prob / total_prob for prob in probabilities]
    cumulative_prob = 0
    r = random.random()  # 随机数0到1之间

    for i, prob in enumerate(probabilities):
        cumulative_prob += prob
        if r < cumulative_prob:
            return kmer_list[i]

# 计算motif的得分
def score(motifs):
    consensus = ""
    k = len(motifs[0])
    for i in range(k):
        count = {'A': 0, 'C': 0, 'G': 0, 'T': 0}
        for motif in motifs:
            count[motif[i]] += 1
        consensus += max(count, key=count.get)
    
    total_score = 0
    for motif in motifs:
        total_score += sum(1 for i in range(k) if motif[i] != consensus[i])
    
    return total_score

# 随机选择motifs
def random_motifs(dna, k):
    motifs = []
    for sequence in dna:
        rand_start = random.randint(0, len(sequence) - k)
        motifs.append(sequence[rand_start:rand_start + k])
    return motifs

# Gibbs Sampler算法
def gibbs_sampler(dna, k, t, N):
    motifs = random_motifs(dna, k)
    best_motifs = motifs[:]
    
    for _ in range(N):
        i = random.randint(0, t - 1)
        motifs_except_i = motifs[:i] + motifs[i + 1:]
        profile = make_profile_with_pseudocounts(motifs_except_i)
        motifs[i] = profile_randomly_generated_kmer(dna[i], k, profile)
        
        if score(motifs) < score(best_motifs):
            best_motifs = motifs[:]
    
    return best_motifs

# 从文件读取数据并运行算法
def main():
    # 文件读取
    with open("rosalind_ba2g.txt") as fp:
        result = fp.readlines()
        k, t, N = map(int, result[0].split())  # k为k-mer长度，t为序列数，N为迭代次数
        dna = [line.strip() for line in result[1:]]  # 读取DNA序列
    
    # 运行Gibbs Sampler 20次，并返回得分最低的motifs
    best_motifs = None
    best_score = float('inf')
    
    for _ in range(20):  # 20次随机启动
        motifs = gibbs_sampler(dna, k, t, N)
        current_score = score(motifs)
        if current_score < best_score:
            best_motifs = motifs
            best_score = current_score
    
    # 输出最优的motifs
    for motif in best_motifs:
        print(motif)

if __name__ == "__main__":
    main()
```

```python
import random
import numpy

def hamming_distance(p, q):
    dist = 0
    for i in range(len(p)):
        if p[i] != q[i]:
            dist += 1
    return dist

def random_initiate(dna, k):
    motif = []
    for i in range(len(dna)):
        j = random.randrange(0, len(dna[i]) - k + 1)
        motif.append(dna[i][j:j+k])
    return motif

def get_score(profile, motif, k):
    dic = {0: 'A', 1: 'C', 2: 'G', 3: 'T'}
    consensus = ''
    for i in range(k):
        j = 0
        most_index, most = 0, 0
        while j < 4:
            if most < profile[j][i]:
                most, most_index = profile[j][i], j
            j += 1
        consensus += dic[most_index]
    score = 0
    for i in range(len(motif)):
        score += hamming_distance(consensus, motif[i])
    return score

def get_profile(motif, k):
    dic = {'A': 0, 'C': 1, 'G': 2, 'T': 3}
    profile = [[1 for i in range(k)] for j in range(4)]
    for i in range(len(motif)):
        for j in range(k):
            profile[dic[motif[i][j]]][j] += 1
    for i in range(len(profile)):
        for j in range(len(profile[0])):
            profile[i][j] = float(format(profile[i][j] / float(len(motif) + 4), '.3f'))
    return profile

def gibbs_sample_motif(text, k, profile):
    dic = {'A': 0, 'C': 1, 'G': 2, 'T': 3}
    prob = []
    for i in range(len(text) - k + 1):
        prob_i = 1
        for j in range(k):
            prob_i *= profile[dic[text[i+j]]][j]
        prob.append(prob_i)
    prob_total = sum(prob)
    for i in range(len(prob)):
        prob[i] = float(format(float(prob[i]) / float(prob_total)))
    j = numpy.random.choice(range(len(text) - k + 1), p=prob)
    return text[j:j+k]

def gibbs_sample_search(dna, k, t, n):
    motif = random_initiate(dna, k)
    best_motif = motif
    profile = get_profile(best_motif, k)
    best_score = get_score(profile, best_motif, k)
    for i in range(n):
        random_number = random.randrange(0, t)
        text = dna[random_number]
        del motif[random_number]
        gibbs_profile = get_profile(motif, k)
        gibbs_motif = gibbs_sample_motif(text, k, gibbs_profile)
        motif.insert(random_number, gibbs_motif)
        profile = get_profile(motif, k)
        score = get_score(profile, motif, k)
        if score < best_score:
            best_score = score
            best_motif = motif
    return best_motif, best_score

with open('rosalind_ba2g.txt', 'r') as fbj:
    lines = fbj.readlines()
    k, t, n = map(int, lines[0].strip().split(' '))
    dna = []
    for i in range(1, len(lines)):
        dna.append(lines[i].strip())
    best_motif = []
    best_score = float('inf')

for i in range(20):
    motif, score = gibbs_sample_search(dna, k, t, n)
    if best_score > score:
        best_score = score
        best_motif = motif

for kmer in best_motif:
    print(kmer)
```
